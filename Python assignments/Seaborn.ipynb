#### 1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

**Ans:**  
1. **scatterplot(): Draws a scatter plot with given data.The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters.**
2. **lineplot(): Draws a line plot with given x and y data.The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters.**
3. **relplot(): Figure-level interface for drawing relational plots onto a FacetGrid.It draws a lineplot or a scatterplot by using the "kind" parameter.**
4. **displot() : Figure-level interface for drawing distribution plots onto a FacetGrid.It draws the histogram/distribution of given column.It has parameters like kind ,color ,etc.**
5. **jointplot() : this plot does both bivariate and univariate anlysis of the x and y columns by providing the scatter plot and the histograms of both the x and y.**

#### 2.  Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data = fmri , x = "timepoint" , y = "signal" , hue = "region" , style= "event")
plt.title("timepoint vs signal")
plt.show()
```

#### 3.  Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
#Plot1
sns.boxplot(data= titanic ,x= 'pclass',y='age')
plt.title("pclass vs age")
plt.show()
```

```python
#Plot2
sns.boxplot(data= titanic ,x= 'pclass',y='fare')
plt.title("pclass vs fare")
plt.show()
```

#### 4.  Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds.head()
```

```python
sns.displot(data= diamonds , x = "price" , hue = "cut")
plt.title("Price histogram")
plt.show()
```

```python
#we can also do this
sns.displot(data= diamonds , x = "price" , hue = "cut",col="cut")
plt.title("Price histogram")
plt.show()
```

#### 5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
#paiplot
sns.pairplot(data = iris, hue = "species")
plt.show()
```

#### 6.  Use the "flights" dataset from seaborn to plot a heatmap

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
#creating a pivot table to create heatmap
pivot_flights = flights.pivot(index = "month" , columns="year",values="passengers")
```

```python
pivot_flights
```

```python
sns.heatmap(pivot_flights)
plt.title("Heatmap")
plt.show()
```
