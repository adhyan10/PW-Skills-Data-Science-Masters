## Matplotlib

#### 1.  What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

**Ans: Matplotlib is a python library used for creating visualizations such as charts,graphs and plots.It provides a vast variety of tools to create quality visualizations which are used in the data anlysis/analytics,scientific computing and machine learning tasks.**  
**Following are the five types of plots that can be plotted using pyplot module of matplotlib:**  
1. **plot(): plots a basic line graph with x and y axis.**
2. **bar(): plots a bar graph with on categorical value and one numerical value.**
3. **scatter(): plots a scatter plot with both values creating dots or markers**
4. **hist(): plots a histogram for a given data.**
5. **pie(): plots a pie charts for the given numeric data.**

#### 2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

**Ans: Scatter plot is a plot where both the values form the co ordinates of the dots(markers) that are plotted in the final resulting plot.**  

```python
import numpy as np
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,c="#13A9EF",marker = "*",alpha=0.8)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot : X vs Y")
plt.grid()
plt.show()
```

#### 3.  Why is the subplot() function used? Draw four line plots using the subplot() function.

**Ans: We use the subplot() when we want to plot/draw multiple plots in a single figure.**     
**The subplot() function takes three arguments that describes the layout of the figure.The layout is organized in rows and columns, which are represented by the first and second argument.The third argument represents the index of the current plot.**

```python
#line 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("plot 1")

#line 2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("plot 2")

#line 3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("plot 3")

#line 4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("plot 4")

plt.suptitle("4 Subplots")
plt.show()
```

#### 4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

**Ans: bar() is used to create bar plots in pyplot class in matplotlib module.Bar plots are used when we want to visualize a categorical value data against a numerical data.**

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color = "#1840E5")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company vs Profit")
plt.grid()
plt.show()
```

```python
plt.barh(company,profit,color = "#1840E5")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company vs Profit")
plt.grid()
plt.show()
```

#### 5. What is a box plot? Why is it used? Using the following data plot a box plot.

**Ans: A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median.The dots after the max and min lines notify the errors in the data. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution**

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
```

```python
plt.boxplot(data)
plt.title("Box plot")
plt.xlabel("box plots")
plt.ylabel("values")
plt.xticks([1,2],["box1","box2"])
plt.grid()
plt.show()
```
